```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("data.csv").dropna()

math = data["Math"]
read = data["Reading"]
write = data["Writing"]
```

```python

m = len(math)
x0 = np.ones(m)
X = np.array([x0, math, read]).T
Y = np.array(write)

# Initial Coefficients
B = np.array([0, 0, 0])
alpha = 0.0001

```

```python
# Cost function
def cost_function(X, Y, B):
    m = len(Y)
    return np.sum((X.dot(B) - Y) ** 2) / (2 * m)

print("Initial Cost:", cost_function(X, Y, B))

# Gradient Descent
def gradient_descent(X, Y, B, alpha, iterations):
    m = len(Y)
    for _ in range(iterations):
        h = X.dot(B)
        loss = h - Y
        gradient = X.T.dot(loss) / m
        B = B - alpha * gradient
    return B
```

```python

# Train the model
newB = gradient_descent(X, Y, B, alpha, 100000)
print("New Coefficients:", newB)
print("Final Cost:", cost_function(X, Y, newB))

# Predictions
Y_pred = X.dot(newB)

# Use sklearn model for R²
model = LinearRegression()
model.fit(data[["Math", "Reading"]], data["Writing"])
print("R² Score:", model.score(data[["Math", "Reading"]], data["Writing"]))
```
