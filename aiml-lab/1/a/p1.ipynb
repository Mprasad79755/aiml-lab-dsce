```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('headbrain.csv')
df.head()
```

```python
x = df['Head Size(cm^3)']
y = df['Brain Weight(grams)']
```

```python
mean_x = np.mean(x)
print(mean_x)
```

```python
mean_y = np.mean(y)
print(mean_y)
```

```python
n = len(x)
n
```

```python
numerator = np.sum((x - mean_x ) * (y-mean_y))
numerator
```

```python
denominator = np.sum((x - mean_x) **2)
denominator
```

```python
b1 = numerator / denominator
b1
```

```python
b0 = mean_y - (b1 * mean_x)
b0
```

```python
x_line = np.linspace(x.min() , x.max(), 100)
y_line = b0+b1* x_line
```

```python
plt.scatter(x,y,color = 'green', label = 'good')
plt.plot(x_line, y_line, color='orange')
```

```python
y_pred = b0+b1*x
ss_res = np.sum((y - y_pred)**2)
ss_tot = np.sum((y - mean_y)**2)
```

```python
r2 = 1-(ss_res / ss_tot)
r2
```
