```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('headbrain.csv')
df.head()
```

```python
x = df['Head Size(cm^3)']
x
```

```python
y = df['Brain Weight(grams)']
y
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
n = len(x)

mean_x, mean_y,n
```

```python
num = np.sum((x-mean_y) * (y - mean_y))
den = np.sum((x-mean_x)**2)
```

```python
b1 = num/den
b1
```

```python
b0 = mean_y -(b1 * mean_x)
b0
```

```python
x_line = np.linspace(x.min() , x.max(),100)
y_line = b0+b1*x_line
```

```python
plt.scatter(x,y,color='green')
plt.plot(x_line, y_line, color="red")
```

```python
y_pred = b0+b1*x
```

```python
ss_res = np.sum((y-y_pred)**2)
ss_tot = np.sum((y-mean_y)**2)
```

```python
r2 = 1-(ss_res/ss_tot)
r2
```
